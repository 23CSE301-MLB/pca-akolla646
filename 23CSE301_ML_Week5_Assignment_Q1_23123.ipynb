```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

df = pd.read_csv("WineQT.csv")
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data)
```

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
```

```python
pca_df = pd.DataFrame(pca_data)
print(pca.explained_variance_ratio_)
```

```python
print("\nExplained Variance (Eigen values):")
print(pca.explained_variance_)
print("\nPrincipal Axes (components/Eigen Vectors):")
print(pca.components_)
print("\nMean of each feature before transformation:")
print(pca.mean_)
```

```python
sns.pairplot(pca_df)
plt.suptitle("Pairplot of Original Features", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(pca_df[0], pca_df[1], c='aqua', alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - 2D Projection')
plt.grid(True)
plt.show()
```

```python
explained_var = pca.explained_variance_ratio_
plt.figure(figsize=(8,5))
plt.bar(range(1, 3), explained_var, alpha=0.7, align='center', label='Individual explained variance')
plt.step(range(1, 3), np.cumsum(explained_var), where='mid', label='Cumulative explained variance',
color='red')
plt.xlabel('Principal Component Index')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Principal Components')
plt.legend()
plt.grid(True)
plt.show()
```
